```python
from sklearn_pandas import DataFrameMapper
from marc_dataframes import marc, zephyr
from sklearn.feature_extraction.text import TfidfVectorizer

columns = marc.select(
    marc.EditionImprint.PUBLICATION_DISTRIBUTION_IMPRINT
)

json = [
    zephyr.load_from_api('txu.059173018498868'),
    zephyr.load_from_api('gri.ark:/13960/t91875m5j'),
    zephyr.load_from_api('mdp.39015022391018')
]

data = zephyr.ZephyrTransformer(columns).transform(json)

mapper = DataFrameMapper(
    list(map(lambda c: (c, TfidfVectorizer(
        ngram_range = (1, 2)
    )), columns)),
    df_out = True
)

X = mapper.fit_transform(data.copy())
display(X)
```
